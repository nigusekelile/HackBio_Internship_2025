```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
# Load the dataset
df = pd.read_csv('nhanes.csv')
df.head()
```

```python
# Process NA values by dropping rows with any NA
df_clean = df.dropna()
df_clean.head()
```

```python
# Visualize distributions
plt.figure(figsize=(15, 10))
```

```python
# BMI Distribution
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
sns.histplot(df_clean['BMI'], kde=True)
plt.title('BMI Distribution')
```

```python
# Weight Distribution (kg)
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 2)
sns.histplot(df_clean['Weight'], kde=True)
plt.title('Weight (kg) Distribution')
```

```python
# Weight in Pounds (Weight * 2.2)
plt.figure(figsize=(10, 8))

df_clean['Weight_pounds'] = df_clean['Weight'] * 2.2
plt.subplot(2, 2, 3)
sns.histplot(df_clean['Weight_pounds'], kde=True)
plt.title('Weight (pounds) Distribution')
```

```python
# Age Distribution
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 4)
sns.histplot(df_clean['Age'], kde=True)
plt.title('Age Distribution')

plt.tight_layout()
plt.show()
```

```python
# Mean 60-second pulse rate
mean_pulse = df_clean['Pulse'].mean()
print(f"Mean 60-second pulse rate: {mean_pulse:.5f}")
```

```python
# Range of diastolic blood pressure (BPDia)
min_bpdia = df_clean['BPDia'].min()
max_bpdia = df_clean['BPDia'].max()
print(f"Diastolic BP range: {int(min_bpdia)}-{int(max_bpdia)}")
```

```python
# Variance and standard deviation of Income
income_var = df_clean['Income'].var()
income_std = df_clean['Income'].std()
print(f"Income variance: {income_var:.2f}, standard deviation: {income_std:.2f}")
```

```python
# Scatter plots for Weight vs Height with coloring
plt.figure(figsize=(18, 5))

# By Gender
plt.subplot(1, 3, 1)
sns.scatterplot(x='Weight', y='Height', hue='Gender', data=df_clean)
plt.title('Weight vs Height (Colored by Gender)')
```

```python
# By Diabetes
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 2)
sns.scatterplot(x='Weight', y='Height', hue='Diabetes', data=df_clean)
plt.title('Weight vs Height (Colored by Diabetes)')
```

```python
# By SmokingStatus
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 3)
sns.scatterplot(x='Weight', y='Height', hue='SmokingStatus', data=df_clean)
plt.title('Weight vs Height (Colored by Smoking Status)')

plt.tight_layout()
plt.show()
```

```python
# T-tests
# Age vs Gender
male_age = df_clean[df_clean['Gender'] == 'male']['Age']
female_age = df_clean[df_clean['Gender'] == 'female']['Age']
t_stat, p_value = ttest_ind(male_age, female_age)
print(f"\nT-test (Age vs Gender): p-value = {p_value:.5f}")
```

```python
from scipy.stats import ttest_ind

# Filter data by gender and age
male_age = df_clean[df_clean['Gender'] == 'male']['Age'].dropna()
female_age = df_clean[df_clean['Gender'] == 'female']['Age'].dropna()

# Check if both groups have data
if len(male_age) == 0 or len(female_age) == 0:
    print("One or both groups have no data. Cannot perform t-test.")
else:
    # Perform t-test
    t_stat, p_value = ttest_ind(male_age, female_age, equal_var=False)
    print(f"\nT-test (Age vs Gender): p-value = {p_value:.5f}")
```

```python
# BMI vs Diabetes
diabetes_bmi = df_clean[df_clean['Diabetes'] == 'Yes']['BMI']
no_diabetes_bmi = df_clean[df_clean['Diabetes'] == 'No']['BMI']
t_stat, p_value = ttest_ind(diabetes_bmi, no_diabetes_bmi)
print(f"T-test (BMI vs Diabetes): p-value = {p_value:.5f}")
```

```python
# AlcoholYear vs RelationshipStatus (example: Committed vs Single)
committed = df_clean[df_clean['RelationshipStatus'] == 'Committed']['AlcoholYear']
single = df_clean[df_clean['RelationshipStatus'] == 'Single']['AlcoholYear']
t_stat, p_value = ttest_ind(committed, single)
print(f"T-test (AlcoholYear vs RelationshipStatus): p-value = {p_value:.5f}")
```
